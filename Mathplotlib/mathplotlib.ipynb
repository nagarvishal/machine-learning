##### # from pandas dataframe - 

```python

batsman = pd.read_csv("files/sharma-kohli.csv")
```

```python
batsman
```

```python
plt.plot(batsman["index"],batsman["V Kohli"])
```

##### Plotting multiple plots - 

```python
plt.plot(batsman["index"],batsman["V Kohli"])
plt.plot(batsman["index"],batsman["RG Sharma"])

plt.title("Rohit Shrama Vs Virat Kohali Career Comparision")
plt.xlabel("Seasion IPL")
plt.ylabel("Runs Scored")
```

##### Colors(hex) and line(width and style) and marker(size)

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="#0074ab",linestyle="solid",linewidth=2)
plt.plot(batsman["index"],batsman["RG Sharma"],color="#ab003e",linestyle="dashed",linewidth=3)

plt.title("Rohit Shrama Vs Virat Kohali Career Comparision")
plt.xlabel("Seasion IPL")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="#0074ab",linestyle="solid",linewidth=2,marker='D',markersize=10)
plt.plot(batsman["index"],batsman["RG Sharma"],color="#ab003e",linestyle="dashed",linewidth=3,marker='o')

plt.title("Rohit Shrama Vs Virat Kohali Career Comparision")
plt.xlabel("Seasion IPL")
plt.ylabel("Runs Scored")
```

##### legend -> location

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="#0074ab",linestyle="solid",linewidth=2,marker='D',markersize=10,label="Virat")
plt.plot(batsman["index"],batsman["RG Sharma"],color="#ab003e",linestyle="dashed",linewidth=3,marker='o',label="Rohit")

plt.title("Rohit Shrama Vs Virat Kohali Career Comparision")
plt.xlabel("Seasion IPL")
plt.ylabel("Runs Scored")
plt.legend(loc="upper left")
```

##### Limiting axes

```python
price = [48000,54000,57000,49000,47000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2017,2019)
```

##### Grid - 

```python
plt.plot(batsman["index"],batsman["V Kohli"],color="#0074ab",linestyle="solid",linewidth=2,marker='D',markersize=10,label="Virat")
plt.plot(batsman["index"],batsman["RG Sharma"],color="#ab003e",linestyle="dashed",linewidth=3,marker='o',label="Rohit")

plt.title("Rohit Shrama Vs Virat Kohali Career Comparision")
plt.xlabel("Seasion IPL")
plt.ylabel("Runs Scored")
plt.legend(loc="upper left")
plt.grid()
```
