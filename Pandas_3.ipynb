```python
# find the last match played by virat kohali in delhi
ipl.head()
ipl["all_players"] = ipl["Team1Players"] + ipl["Team2Players"]
```

```python
def isVirathKohali(players_list):
    return 'V Kohli' in players_list
```

```python
ipl["is_virat_play"] = ipl["all_players"].apply(isVirathKohali)
ipl.head()
```

```python
ipl[(ipl["City"] == "Delhi") & (ipl["is_virat_play"] == True)].drop_duplicates(subset=["City","is_virat_play"])
```
