# More Important Functions

```python
import numpy as np
import pandas as pd
```

```python
#value_counts() - (series and dataframe)

a = pd.Series([1,1,1,2,2,3]);
print(a.value_counts())

marks = pd.DataFrame([
    [100,80,10],
    [90,70,7],
    [120,100,14],
    [80,70,14],
    [80,70,14]
],columns=['iq','marks','package'])

marks
marks.value_counts()
```

```python
ipl = pd.read_csv('ipl-matches.csv');
ipl.head(2)
```

```python
#Find which player has won most player of the match -> in finals and quterfinals.
ipl[~ipl["MatchNumber"].str.isdigit()]["Player_of_Match"].value_counts()
```

```python
# Toss Disition Plot -
ipl["TossDecision"].value_counts().plot(kind="pie")
```

```python
# How many matches each team player -
ipl.head()
ipl["Team1"].value_counts()
ipl["Team2"].value_counts()
```

```python
pd.concat([ipl["Team1"], ipl["Team2"]], axis=0).value_counts()
(ipl["Team1"].value_counts()+ipl["Team2"].value_counts()).sort_values(ascending=False)
```

```python
# sort_values() - ( Series and DataFrame) -> ascening , na_position, inplace, multiple_cols
x = pd.Series([4,5,2,11,7,5,6])
x.sort_values()
x.sort_values(ascending=True,inplace=True)
x
```

```python
movies = pd.read_csv("movies.csv")
movies.head(3)
```

```python
movies.sort_values("title_x",ascending=False)
```

```python
students = pd.DataFrame(
    {
        'name':['nitish','ankit','rupesh',np.nan,'mrityunjay',np.nan,'rishabh',np.nan,'aditya',np.nan],
        'college':['bit','iit','vit',np.nan,np.nan,'vlsi','ssit',np.nan,np.nan,'git'],
        'branch':['eee','it','cse',np.nan,'me','ce','civ','cse','bio',np.nan],
        'cgpa':[6.66,8.25,6.41,np.nan,5.6,9.0,7.4,10,7.4,np.nan],
        'package':[4,5,6,np.nan,6,7,8,9,np.nan,np.nan]

    }
)

students
```

```python
students.sort_values("name")
students.sort_values("name",ascending=False)
students.sort_values("name",ascending=False,na_position="first")
```

```python
# Sorting on Multiple Columns -
movies.head(3)
movies.sort_values(["year_of_release","title_x"],ascending=True)
movies.sort_values(["year_of_release","title_x"],ascending=[True,False])
```
