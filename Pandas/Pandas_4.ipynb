##### agg method - passing dict

```python
genres.sum(numeric_only=True)
genres.agg(
    {
        "Runtime" : "mean",
        "IMDB_Rating" : "mean",
        "No_of_Votes" : "sum",
        "Gross" : "sum",
        "Metascore" : "min"
    }
)
```

```python
genres.agg(["min","max","mean"])
```

```python
genres.agg(
    {
        "Runtime" : ["mean","max","min"],
        "IMDB_Rating" : "mean",
        "No_of_Votes" : ["sum","max"],
        "Gross" : "sum",
        "Metascore" : "min"
    }
)
```
