```python
# find all the final winners
final_mathces = ipl["MatchNumber"]=="Final"
new_df = ipl[final_mathces]
new_df[["Season","WinningTeam"]]
ipl[ipl["MatchNumber"]=="Final"][["Season","WinningTeam"]]
```

```python
# How Many Super Over finishis Occured
ipl.head(5)
ipl[ipl["SuperOver"]=="Y"][["WinningTeam"]]
```

```python
# How Many Mathced has CSK won in kolkata -
ipl.head(10)
ipl[(ipl["City"] == "Kolkata") & (ipl["WinningTeam"] == "Chennai Super Kings")].shape[0]
```

```python
# Toss Winner is Mathch Winner Percentage -
ipl.head()
TWP = (ipl[ipl["TossWinner"] == ipl["WinningTeam"]].shape[0]/ipl.shape[0])*100
TWP
```

```python
# Movies with rating Higher than 8 and votes > 10000
movies.head()
movies[(movies["imdb_rating"] > 8) & (movies["imdb_votes"] > 10000)].shape[0]
```

```python
# Action Movies with rating higher then 7.5
movies.head(10)
# movies[(movies["genres"].str.contains("Action")) & (movies["imdb_rating"] > 7.5)].shape[0]
movies[(movies["genres"].str.split("|").apply(lambda x: 'Action' in x )) & (movies["imdb_rating"] > 7.5)].shape[0]
```

```python
# Write a function that can return the track record of 2 team against each other
def func(first_team,second_team):
    dm = ipl[(((ipl["Team1"] == first_team) & (ipl["Team2"] == second_team)) | ((ipl["Team2"] == first_team) & (ipl["Team1"] == second_team)))]
    return dm

func("Rajasthan Royals","Punjab Kings")
```
