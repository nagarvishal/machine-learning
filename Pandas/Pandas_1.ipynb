# Pandas Series -
A Pandas Series like a column in a table. It is a one-d array holding data of any type.

```python
import numpy as np
import pandas as pd
```

# Series From List

```python
#string
country = ['India','Pakasthin','USA','Nepal','shrilanka']
pd.Series(country)
```

```python
#Integer
runs = [56,34,27,89,12,34,56,100]
runs_series = pd.Series(runs)
```

```python
# Custome Index
marks = [56,78,34,56,89,100]
subjects = ["hindi","english","mathematics","science","physics","chemystry"]
pd.Series(marks, index=subjects)


```

```python
#Setting A Name -
marks = pd.Series(marks, index=subjects, name="Vishal Ke Marks")
marks
```

# Series From DICT -

```python
marks = {
    'maths':67,
    'english':89,
    'science':34,
    'hindi':34
}

marks_series = pd.Series(marks,name="Vishal Ke Marks")
marks_series
```

# Some Important Series Methods

```python
movies = pd.read_csv('/content/bollywood.csv',index_col="movie").squeeze()
vk = pd.read_csv('/content/kohli_ipl.csv',index_col='match_no').squeeze()
subs = pd.read_csv('/content/subs.csv').squeeze()
```

```python
# astype
import sys
vk.head()
print(sys.getsizeof(vk))
print(sys.getsizeof(vk.astype('int16')))
```

```python
# between
vk.head()
vk[vk.between(51,99)].size
```

```python
#clip
subs.clip(100,200)
```

```python
# drop_duplicates
temp = pd.Series([1,1,2,2,3,3,3])
temp.drop_duplicates()
temp.drop_duplicates(keep="last")
movies.drop_duplicates()
movies.drop_duplicates(keep="last")
```

```python
temp.duplicated()
temp.duplicated().sum()
vk.duplicated().sum()
```

```python
# isnull
temp = pd.Series([1,2,3,np.nan, 5,6, np.nan, np.nan, 8, 10])
temp
temp[~temp.isnull()]
vk[~vk.isnull()]
```

```python
print(temp.size)
print(temp.count())
```

```python
# dropna -> function use for removing missing values
temp.dropna()
vk.dropna()
```

```python
# fillna
temp
temp.fillna(temp.mean())
```
